```python
# Since the is a regression problem statement we will not check for imbalanced dataset
# feature Scaling
# STep1 segregate dependent and independent variables
X=dataset.iloc[:, :-1] #independent vars
y=dataset.iloc[:,-1]
```

```python
X.shape
```

```python
y
```

```python
#Step 2 Train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
# Random State Task 
```

```python
X_train.shape
## Rows - 404
## Columns - 13
```

```python
X_test.shape 
```

```python
y_train
```

```python
# Applying Standard scalar (Normalization ? Standardization)
from sklearn.preprocessing import StandardScaler
Scaler= StandardScaler()
```

```python
X_train=Scaler.fit_transform(X_train)
X_test=Scaler.transform(X_test)
```

```python
X_train
```

```python
# Model Building 
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
#training the model
regression.fit(X_train,y_train)
```

```python
#y=mx+c
# C is intercept
print(regression.intercept_)
```

```python
# finding coefficients 
print(regression.coef_)
```

```python
# Testing data 
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test,reg_pred)
```

```python
# Checking Accuracy 
# R - Square value 
from sklearn.metrics import r2_score
accuracy_score=r2_score(y_test,reg_pred)
```

```python
accuracy_score
```
