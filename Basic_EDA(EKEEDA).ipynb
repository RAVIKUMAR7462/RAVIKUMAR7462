```python
IQR = d['Salaries'].quantile(0.75) - d['Salaries'].quantile(0.25)
lower_limit = d['Salaries'].quantile(0.25) - (IQR * 1.5)
upper_limit = d['Salaries'].quantile(0.75) + (IQR * 1.5)
IQR 
```

```python
lower_limit
```

```python
upper_limit 
```

```python
# triamming
outliers_d = np.where(d['Salaries'] > upper_limit, True, np.where(d['Salaries'] < lower_limit, True, False))
sum(outliers_d)
```

```python
d.shape
```

```python
d_trimmed = d.loc[~(outliers_d), ]
d.shape, d_trimmed.shape
```
